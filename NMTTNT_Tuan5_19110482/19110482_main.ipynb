```python
from sys import maxsize
from itertools import permutations
V = 4


def TSP(graph, s):

    # lưu trữ tất cả các đỉnh ngoài đỉnh gốc
    vertex = []
    for i in range(V):
        if i != s:
            vertex.append(i)

    # lưu trữ trọng lượng tối thiểu Chu kỳ Hamilton
    minpath = maxsize
    nextper = permutations(vertex)
    for i in nextper:

        # lưu trữ trọng số đường dẫn hiện tại (chi phí)
        current = 0

        # tính toán trọng lượng đường dẫn hiện tại
        k = s
        for j in i:
            current += graph[k][j]
            k = j
        current += graph[k][s]

        # cập nhật
        minpath = min(minpath, current)

    return minpath


graph = [[0, 7, 9, 0, 0, 14],
         [7, 0, 10, 15, 0, 0],
         [9, 10, 0, 11, 0, 2],
         [0, 15, 11, 0, 6, 0],
         [0, 0, 0, 6, 0, 9],
         [14, 0, 2, 0, 9, 0]]
print('cost is', TSP(graph, 0))
```
